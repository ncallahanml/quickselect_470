```python
img_height = 1000
img_width = 2000

box_color = (255,255,255)
box_width = 80

gap = 10
n_boxes = 6
left_over = img_width - (n_boxes * box_width + (n_boxes - 1) * gap)
assert left_over >= 0, 'Layout exceeds background width'
```

```python
@dataclass
class Box():
    _width : int
    _center : tuple
    _number : int 
    _color : tuple
    thickness = 2
    font_scale = 1
    temp_color = None
#     color_gen = deepcopy(itertools.cycle([(0,0,255),(0,0,155),(0,0,55),None]))
    
    @property
    def points(self):
        xc, yc = self._center
        width = self._width
        start_point = ((xc - width // 2), (yc - width // 2))
        end_point = ((xc + width // 2), (yc + width // 2))
        return start_point, end_point
    
    @property
    def center(self):
        return self._center
    
    @property
    def width(self):
        return self._width
    
    @property
    def number(self):
        return self._number
    
    @property
    def color(self):
        self.original_color = self.color
        return self._color
    
#     def revert_color(self):
#         self.color = self.original_color
#         return
    
    @center.setter
    def center(self, center : tuple):
        if not isinstance(center, tuple): raise TypeError(f'Center must be a tuple not {type(center)}')
        if not len(center) == 2: raise ValueError('Center must length 2')
        self._center = center
        return 
    
    @number.setter
    def number(self, number : int):
        if not isinstance(number, (int, np.int32, np.int64)): raise TypeError(f'Number must be an integer, not {type(number)}')
        self._number = number
        self.temp_color = (0,0,251)
        return
    
    @color.setter
    def color(self, color : tuple):
        if not isinstance(color, tuple): raise TypeError(f'Color must be a tuple, not {type(number)}')
        if not len(color) == 3: raise ValueError(f'Color must length 3, not {len(color)}')
        for item in color: assert isinstance(item, (int, np.int32, np.int64, np.uint8))
        self._color = color
        return
            
    def move_down(self, gap=10):
#         curr_center = self.center
#         self.center = (curr_center[0], curr_center[1] + gap + self._width)
        self.offset(dy = gap + self._width // 2)
        return
    
    def move_up(self, gap=10):
#         curr_center = self.center
#         self.center = (curr_center[0], curr_center[1] + gap + self._width)
        self.offset(dy = -(gap + self._width // 2))
        return
    
#     def split_up(self, box, left_box=None, left_gap=20, xcmin=0, gap=10):
#         self.move_up(gap=gap)
#         box.move_up(gap=gap)
#         if left_box is None:
#             xcmin = 0
#         elif left_box:
#             if not left_gap: left_gap = gap
#             (_,_),(lxcmax,_) = left_box.points
#             xcmin = lxcmax + left_gap + self.width // 2
#         box1, box2 = self, left_box
#         box1.center = (xcmin, box1.center[1])
#         box2.center = (xcmin + box2.width + gap, box2.center[1])
#         return
    
#     def merge_down(self, box, left_box=None, left_gap=20, xcmin=0, gap=10):
#         self.move_up(gap=gap)
#         box.move_up(gap=gap)
#         if left_box is None:
#             xcmin = 0
#         elif left_box:
#             if not left_gap: left_gap = gap
#             (_,_),(lxcmax,_) = left_box.points
#             xcmin = lxcmax + left_gap + self.width // 2
#         box1, box2 = self, left_box
#         box1.center = (xcmin, box1.center[1])
#         box2.center = (xcmin + box2.width + gap, box2.center[1])
#         return
        
    def offset(self, dx=0, dy=0):
        self.center = (self.center[0] + dx, self.center[1] + dy)
        return
    
    @staticmethod
    def align_horizontal(boxes, separation=None, xcmin=0, gap=10):
        separation = boxes[0].width // 2 if separation is None else separation
        xc = -gap - separation // 2 + xcmin
        for box in boxes:
            xc += gap + separation
            box.center = (xc, box.center[1])
        return
            
    @staticmethod
    def align_left(boxes, separation=None, xcmin=0, gap=10):
        Box.align_horizontal(boxes, separation=separation, xcmin=xcmin, gap=gap)
        return

    @staticmethod
    def align_right(boxes, separation=None, xcmax=0, gap=10):
        separation = -boxes[0].width // 2 if separation is None else -separation
        #### this is probably wrong
        Box.align_horizontal(boxes[::-1], separation=-separation, xcmin=xcmax + 2*gap + 2*separation, gap=gap)
        return
    
    def swap(self, box):
        self.center, box.center = box.center, self.center 
        return
        
    def graph(self, img):
        start_point, end_point = self.points
        font = cv2.FONT_HERSHEY_SIMPLEX
        if self.temp_color:
            cv2.rectangle(img, start_point, end_point, self.temp_color, -1)
#             self.temp_color = next(self.color_gen)
            if self.temp_color == (0,0,251): self.temp_color = (0,0,250)        
            elif self.temp_color == (0,0,250): self.temp_color = (0,0,175)
            elif self.temp_color == (0,0,175): self.temp_color = (0,0,100)
            elif self.temp_color == (0,0,100): self.temp_color = None
        else:
            cv2.rectangle(img, start_point, end_point, self._color, self.thickness)
        if isinstance(self._number, (str, int, np.int32, np.int64)):
            text = str(self._number)
            text_size = cv2.getTextSize(text, font, self.font_scale, self.thickness)[0]
            text_x = self.center[0] - text_size[0] // 2
            text_y = self.center[1] + text_size[1] // 2
            cv2.putText(img, text, (text_x, text_y), font, self.font_scale, self._color, self.thickness, cv2.LINE_AA)
        return img
    
```

```python
def partition_gen(arr, left, right):
    x = arr[right]
    i = left
    indices = list()
    for j in range(right, left, -1):
        if arr[j] <= x:
            indices.append((i, j))
            arr[i], arr[j] = arr[j], arr[i]
            i += 1
#             yield i - 1, j
    return indices

# def partition_gen(arr, left, right):
# #     x = arr[right]
#     i = 0
#     j = right
#     indices = list()
#     while i != k:
#         if arr[j] <= x:
#             indices.append((i, j))
#             arr[i], arr[j] = arr[j], arr[i]
#             i += 1
#         elif arr[i] <= x:
#             pass
# #             yield i - 1, j
#     return indices
```

```python
def center_split(boxes, xcmin=0, small_gap=10):
    xcmin += boxes[0].width // 2
    for i, box in enumerate(boxes):
        box.move_up()
        box.center = (xcmin, box.center[1])
        xcmin += box.width + small_gap
    return

def recursive_split(boxes, xcmin=0, small_gap=10, large_gap=20, gap_multiple=1):
    if len(boxes) == 1:
        return
    left_boxes, right_boxes = split(boxes, len(boxes) // 2)
    
    center_split(left_boxes, xcmin=xcmin, small_gap=small_gap*gap_multiple)
    graph(boxes)

#     left_len = len(left_boxes)
    right_xcmin = left_boxes[-1].center[0] + left_boxes[-1].width + large_gap
    center_split(right_boxes, xcmin=right_xcmin, small_gap=small_gap)
    graph(boxes)
    
    recursive_split(left_boxes, xcmin=xcmin, gap_multiple=gap_multiple + 1)
    recursive_split(right_boxes, xcmin=right_xcmin, gap_multiple=gap_multiple + 1)
    return

def merge_two(left_boxes, right_boxes):
    i = 0
    j = 0
    indices = list()
    if isinstance(left_boxes, Box): left_boxes = [left_boxes]
    if isinstance(right_boxes, Box): right_boxes = [right_boxes]
    while len(indices) < len(left_boxes) + len(right_boxes):
        if i < len(left_boxes) and j < len(right_boxes):
            while left_boxes[i].number <= right_boxes[j].number:
                indices.append(i)
                if i + 1 == len(left_boxes): break
                i += 1
            while right_boxes[j].number < left_boxes[i].number:
                indices.append(len(left_boxes) - 1 + j) # index in parent array before split
                if j + 1 == len(right_boxes): break
                j += 1
        elif i < len(left_boxes): #could extend indices, but this is more clear?
#             while i < len(left_boxes):
            indices.append(i)
            i += 1
        elif j < len(right_boxes): #could extend indices, but this is more clear?
#             while i < len(right_boxes):
            indices.append(len(left_boxes) - 1 + j)
            j += 1
#         else:
#             break
    assert len(indices) == len(left_boxes) + len(right_boxes)
    return indices

def pairs(boxes):
    return [boxes[i:i+2] if i < len(boxes) - 1 else boxes[-1:] for i in range(0, len(boxes), 2)]
    
def flatten(boxes):
    # only works with integers
    if isinstance(boxes, Box): return [boxes]
    new_list = list()
    for box in boxes:
        new_list.extend(flatten(box))
    return new_list

def center_merge(boxes, small_gap=1, large_gap=2):
    level = 1
    grouped_boxes = boxes
    while len(grouped_boxes) > 1:
#         level *= 2
#         def split_level(boxes, i, level):
#             index = i + level if i + level < len(boxes)
#         grouped_boxes = [[boxes[i:i+level] if i + level < len(boxes) else boxes[i:]] for i in range(0, len(boxes)+1, level)]
        temp_grouped_boxes = list()    
        for pair in pairs(grouped_boxes):
            if len(pair) == 1: 
                assert pair[0] is grouped_boxes[-1]
                temp_grouped_boxes.append(pair)
                break
            left_boxes, right_boxes = pair
            indices = merge_two(flatten(left_boxes), flatten(right_boxes))
#             left_boxes.extend(right_boxes)
            full_boxes = flatten(left_boxes) + flatten(right_boxes)
#             assert len(full_boxes) == len(left_boxes) + len(right_boxes), f'{full_boxes}, \n\n{left_boxes}, \n\n{right_boxes}'
            full_boxes = [full_boxes[i] for i in indices]
            assert len(full_boxes) == len(flatten(left_boxes)) + len(flatten(right_boxes))
            temp_grouped_boxes.append(full_boxes)
        grouped_boxes = temp_grouped_boxes
        flat_boxes = flatten(grouped_boxes)
        for box in flat_boxes:
            box.move_down()
        xcmin = 0
        for grouped_box in grouped_boxes:
            flat_group = flatten(grouped_box)
            Box.align_left(flat_group, xcmin=xcmin, gap=small_gap)
            xcmin = flat_group[-1].center[0] + flat_group[-1].width + large_gap
        graph(flatten(grouped_boxes))
    return

def split(lis, index):
    left_list = lis[:index]
    right_list = lis[index:]
    assert len(left_list) + len(right_list) == len(lis)
    return left_list, right_list

def merge_boxes(nums):
    xc = box_width // 2 + 10
    boxes = list()
    for num in nums:
        yc = img_height // 2
        box = Box(box_width, (xc, yc), num, box_color)
        xc += gap + box_width
        boxes.append(box)
#         img = box.graph(img)
    #### above is WET
    graph(boxes)
    recursive_split(boxes) # just handles animation
    boxes = [[box] for box in boxes]
    center_merge(boxes)
    print(boxes)
    graph(flatten(boxes))
    return

merge_boxes([10, 13, 4, 3, 1, 23])
```
